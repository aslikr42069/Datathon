```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dtype_options = {
    'Unnamed: 0': 'int64',
    'Flight_ID': 'str',
    'ProductType': 'str',
    'ProductName': 'str',
    'Quantity': 'float64',
    'TotalSales': 'float64',
    'Aeronave': 'str',
    'DepartureStation': 'str',
    'ArrivalStation': 'str',
    'DestinationType': 'str',
    'Origin_Type': 'str',
    'Capacity': 'float64',
    'Passengers': 'float64',
    'Bookings': 'float64'
}

parse_dates = ['STD', 'STA']

flightSales = pd.read_csv(FLIGHT_SALES_PATH, dtype=dtype_options, parse_dates=parse_dates)
flightSales.head()
```

```python
flightSales['STD'] = pd.to_datetime(flightSales['STD'])
flightSales['STA'] = pd.to_datetime(flightSales['STA'])
flightSales.head()
```

```python
flightSales['month'] = flightSales['STD'].dt.to_period('M')
monthFlightSales = flightSales.groupby(['ProductType', 'month']).sum(numeric_only=True)
monthFlightSales.reset_index(inplace=True)
monthFlightSales.head(20)
```

```python
monthFlightSales.columns
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for product_type in monthFlightSales['ProductType'].unique():
    product_df = monthFlightSales[monthFlightSales['ProductType'] == product_type]
    ultimo_mes = product_df['month'].max()
    filtered_df = product_df[product_df['month'] != ultimo_mes]
    ax.plot(filtered_df['month'].astype(str), filtered_df['TotalSales'], label=product_type)

ax.set_title('Total de ventas por mes')
ax.set_xlabel('Mes')
ax.set_ylabel('Total de ventas')
ax.legend()
ax.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
